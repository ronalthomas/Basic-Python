```python
# simple grading system
grade = input ("enter your grade")
grade = float (grade)
if grade >= 90:
    print ("your grade is A")
elif grade >=80:
    print ("your grade is B")
elif grade >=70:
    print ("your grade is c")
elif grade >=60:
    print ("your grade is satisfactory")
else:
    print ("you have failed the exam")
```

```python
# age category checker
age = input ("enter your grade")
age = int (age)
if age >=90:
    print ("you are a pensioner")
elif age >= 50:
    print ("you are retired")
elif age >=18:
    print ("you are an adult")
elif age >=12:
    print ("you are a teenager")
elif age >=7:
    print ("you are a child")
elif age >=5:
    print ("your age is between 7 and 5")
else:
    print ("grow up")
```

```python
#Simple Calculator (2 numbers + operator)
a = float (input ("enter the value for a"))
b = float (input ("enter the value for b"))
operator = input ("enter your operator")
if operator == "+":
    print (a+b)
elif operator == "-":
    print (a-b)
elif operator == "*":
    print (a*b)
elif operator =="/":
    print (a/b)
else:
    print ("invalid operator entered")
```

```python
#Traffic Light System
colour = input ("enter the traffic light colour")
if colour == "red":
    print ("stop now")
elif colour == "orange":
    print ("get ready to stop")
elif colour == "green":
    print ("you may go now")
else:
    print ("invalid colour entered")
```

```python
#dicount calculator
price = input ("enter your price")
price = float (price)
if price >=100:
    discount_price = price *0.10
    new_price = price - discount_price
    print ("the discount is 10 %", "the discount price is", discount_price,"the final price is",new_price)
elif price>=50:
    discount_price = price *0.05
    new_price = price - discount_price
    print ("the discount is 5%","the discount price is",discount_price,"the final price is",new_price)
else:
    print ("no discount available")
```
