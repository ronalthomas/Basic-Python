```python
#project -1 (simple grading system)
grade = input ("enter your grade")
grade = float (grade)
if grade >= 90:
    print ("your grade is A+")
if grade >= 80:
    print ("your grade is A")
if grade >= 70:
    print ("your grade is B+")
if grade >= 60:
    print ("your grade is c")
else:
    print ("you have failed")
    print ("you need to re write your exam")
print ("results finished")
```

```python
#project-2 (age category checker)
age = input ("enter your age")
age = int (age)
if age >= 70:
    print ("you are a pensioner")
else:
    if age >=20:
        print ("you are an adult")
    else:
        if age >=13:
            print ("you are a teenager")
        else:
            if age <= 5:
                if age >=3:
                        print ("you are a child between the age of 3 and 5")
                else:
                        print ("you are a baby")
print ("results finished")
```

```python
#project-3 (ATM simulator) simulate a simple ATM that checks your balance before withdrawal

withdrawal = input ("enter your withdrawal")
withdrawal = float (withdrawal)
current_balance = input ("enter your current balance")
current_balance = float (current_balance)
if withdrawal > current_balance:
    print ("insufficient balance")
else:
    if withdrawal ==0:
        print ("enter a correct amount")
    else:
        print ("your withdrawal is being processed")
print ("finished")
```

```python
#project -4 (Number guessing game)
number = input ("enter your number between 1 and 10")
number = int (number)
secret_number = 7
if number == secret_number:
    print ("your guess is correct, you won the game")
else:
    if number >7:
        print ("the number is a little too big")
    else:
        if number <7:
            print ("the number is a little too small")

    print ("you have two more tries")
print ("finished")
```

```python
#project-5 (discount calculator)
purchase_amount = input ("enter your purchase amount")
purchase_amount = float (purchase_amount)
if purchase_amount < 100:
    print ("no discount applied")
else:
    if purchase_amount >=100:
        if purchase_amount <=499:
         discount = 100 * 0.10
         final_price = purchase_amount - discount
         print ("discount of 10% applied")
         print ("the final price is", final_price)
        else:
            if purchase_amount >=500:
                discount = 100 * 0.50
                final_price = purchase_amount - discount
                print ("discount of 50% applied")
                print ("the final price is", final_price)
print ("proceed to checkout")
```
